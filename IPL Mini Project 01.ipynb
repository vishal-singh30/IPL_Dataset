# Q11. Plot a histogram of number of matches played in a season by players.

```python
import matplotlib.pyplot as plt
import pandas as pd

# Plotting the histogram
plt.hist(data['Mat'],10,color='blue',edgecolor='black')
plt.xlabel('Matches')
plt.ylabel('Frequency')
plt.title('Histogram of no of matches played by player')
plt.show()
```

# Q12. Plot the histogram of balls faced by players.

```python
# Plotting the histogram
plt.hist(data['BF'],10,color='blue',edgecolor='black')
plt.xlabel('Balls Faced')
plt.ylabel('Frequency')
plt.title('Histogram of no of matches played by player')
plt.show()
```

# Q13. Top 20 players with most runs in a season.

```python
data.nlargest(20,'Runs')[['Player','Runs']]
```

# Q14. Bar chart for top 20 players with best strike rate in a season.

```python
Top_20=data.nlargest(20,'SR')[['Player','SR']]
# Plotting the histogram
plt.bar(Top_20['Player'],Top_20['SR'],color='green')
plt.xlabel('Player')
plt.ylabel('Strike Rate')
plt.xticks(rotation=50,ha='right')
plt.show()
```

# Q15. Top 30 players with highest number of sixes.

```python
data.nlargest(30,'6s')[['Player','6s']]
```

# Q16. Scatter plot of runs scored by a player v/s balls faced in a season.

```python
import matplotlib.pyplot as plt
import pandas as pd

plt.figure(figsize=(10,6))
plt.scatter(data['Runs'],data['BF'],marker='*',color='blue')
plt.xlabel('Player')
plt.ylabel('BF')
plt.show()
```

# Q17. KDE plot for averages on a season.

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
sns.kdeplot(data['Avg'],fill='True',color='red')
plt.xlabel('Average')
plt.ylabel('Density')
plt.show()
```
